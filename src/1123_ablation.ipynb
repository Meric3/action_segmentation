```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python

names = list(set(os.listdir('./fig/ablation/1')) & set(os.listdir('./fig/ablation/2'))
             & set(os.listdir('./fig/ablation/3'))& set(os.listdir('./fig/ablation/pca')))
names.sort()


types = ['pca', '3', '2', '1']
    
for i, name in enumerate(names):
    if name == '.ipynb_checkpoints': continue
    images = []
    for t in types:
        path = os.path.join('./fig/ablation', t)
        p = os.path.join(path, name)
        images.append(Image.open(p))
        print(t, name)
#         break
    break
```

```python
fig = plt.figure()

fig.set_figheight(5)
fig.set_figwidth(10)

# Adds subplot on position 1
ax = fig.add_subplot(411)
ax.imshow(images[0])
ax = fig.add_subplot(412)
ax.imshow(images[1])
ax = fig.add_subplot(421)
ax.imshow(images[2])
ax = fig.add_subplot(422)
ax.imshow(images[3])
```

```python
fig, axes = plt.subplots(2, 2)
fig.set_size_inches(20, 20)

for i in range(2):
    for j in range(2):
        axes[i, j].set_xticks([])
#         axes[i, j].set_yticks([])
        ax.set_zticks([])

        axes[i, j].imshow(images[2*i+j], cmap='Greys_r')
        
```
